```python
import pandas as pd
```

```python
df = pd.read_csv("Netflix_movies_and_tv_shows_clustering.csv")

```

```python
print(df.info())
print(df.head())
```

```python
print(df.isnull().sum())  #identify the null values
```

```python
df.dropna(inplace=True)  #drop  rows having missing values
```

```python
df.drop_duplicates(inplace=True) # remove duplicate rows
```

```python
text_columns = df.select_dtypes(include='object').columns
for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
print(df.head())
```

```python
if 'date_added' in df.columns:   #converting date into proper format
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
    df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')
```

```python
print(df.head)
```

```python
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]  #column
```

```python
print(df.head)
```

```python
if 'release_year' in df.columns:
    df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce').astype('Int64')
```

```python
df.to_csv("cleaned_netflix_data.csv", index=False) #saving it
```
